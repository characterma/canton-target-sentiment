```python
import pandas as pd
from pathlib import Path
import os
from sklearn.metrics import confusion_matrix
```

```python
model_dir = Path("../model/20201021_restaurant_bert")
```

```python
df = pd.read_csv(model_dir / "eval_details_dev_epoch0.csv")
```

```python
cm_int = confusion_matrix(df['label'], df['pred'], labels=['neutral', 'negative', 'positive'])
cm_norm_row = confusion_matrix(df['label'], df['pred'], normalize='true', labels=['neutral', 'negative', 'positive'])
cm_norm_col = confusion_matrix(df['label'], df['pred'], normalize='pred', labels=['neutral', 'negative', 'positive'])
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
sn.heatmap(cm_int, annot=True, xticklabels=['neutral', 'negative', 'positive'], yticklabels=['neutral', 'negative', 'positive'])
plt.show()
```

```python
sn.heatmap(cm_norm_row, annot=True, xticklabels=['neutral', 'negative', 'positive'], yticklabels=['neutral', 'negative', 'positive'])
plt.show()
```

```python
sn.heatmap(cm_norm_col, annot=True, xticklabels=['neutral', 'negative', 'positive'], yticklabels=['neutral', 'negative', 'positive'])
plt.show()
```
